#### Write a python program or function that does:
1. loads the stockholm temperature dataset via appropiate numpy functions

2. generate numpy array of the morning temperatures in Fahrenheit temperature scale. Note temperature was initially in Celsius scale.

3. generate a series plot of the morning temperature in Fahrenheit vs time. Label the title, legend and axes accordingly.

4. generate a histrogram of the morning temperature in Fahrenheit to reflect the temperature distribution. Label the title, legend and axes accordingly.

    
    

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["font.size"] = 14.0
```

```python
def to_farenheit(C):
    """Converts a temperature in Celsius to Farenheit."""
    return 32.0 + (9 / 5) * C
```

```python
stockholm_temps = np.genfromtxt("stockholm_td_adj.dat")

year = (
    stockholm_temps[:, 0] + stockholm_temps[:, 1] / 12.0 + stockholm_temps[:, 2] / 365.0
)

T_farenheit = to_farenheit(stockholm_temps[:, 3])
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

ax.plot(year, T_farenheit)
ax.set_title("Morning Temperatures in Stockholm")
ax.set_xlabel("Year")
ax.set_ylabel(r"Temperature [$^\circ$F]")

plt.show()
```

```python
fig2, ax2 = plt.subplots(figsize=(8, 6))

bins = int(np.sqrt(T_farenheit.shape[0]))

ax2.hist(T_farenheit, bins=bins)

ax2.set_title("Morning Temperatures in Stockholm")
ax2.set_xlabel("Temperature [$^\circ$F]")
ax2.set_ylabel("Count")

plt.show()
```
